```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier 
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=11, stratify=y)
```

```python
y_train
```

```python
num = ['patient_id', 'age',	'children',	'sex',	'region_code',	'bmi', 'smoker',	'bill_amount',	'claimed_amount',	'amount_paid',	'duration',	'year_billing']
```

```python
num
```

```python
cat = []
```

```python
df.groupby('insuranceclaim')['insuranceclaim'].count()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
X_resampled.head()
```

```python
y_resampled.groupby('insuranceclaim')['insuranceclaim'].count()
```

```python
y_resampled.head()
```

## Linear Regression

```python
preprocessor = ColumnTransformer(                 #ColumnTransforme is used for scaling, onehodencoding all tranforms in a single function but not seperatly
    transformers = [
        ('num', StandardScaler(), num),
        ('cat', OneHotEncoder(drop = 'first'), cat)
    ],
    remainder = 'drop'
)
```

```python
pipeline_lr = Pipeline([
    ("prep", preprocessor),
    ("clf", LogisticRegression(class_weight="balanced", max_iter=1000))
])
```

```python
pipeline_lr.fit(X_resampled, y_resampled)
```

```python
y_pred_lr = pipeline_lr.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
pipeline_lr.score(X_test, y_test)
```

## Random Forest Classifier

```python
pipeline_rf = Pipeline([
    ("prep", preprocessor),
    ("clf", RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
pipeline_rf.fit(X_resampled, y_resampled)
```

```python
y_pred_rf = pipeline_rf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
pipeline_rf.score(X_test, y_test)
```

## XG Boost

```python
pipeline_xgb = Pipeline([
    ("prep", preprocessor),
    ("clf", XGBClassifier(n_estimators=100, learning_rate=0.1, random_state=42))
])
```

```python
pipeline_xgb.fit(X_resampled, y_resampled)
```

```python
y_pred_xgb = pipeline_xgb.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_xgb))
```

```python
pipeline_xgb.score(X_test, y_test)
```
